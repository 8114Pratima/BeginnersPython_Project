```python
# Mapping approach
df["sentiment"]=df["sentiment"].map({"neutral":0,"negative":1,"positive":2})
```

```python
df["sentiment"].value_counts(normalize=True)
```

```python
# split the data in independent and dependent variables
```

```python
X=df["text"]
Y=df["sentiment"]
```

```python
X.head()
```

```python
Y.head()
```

```python
# preprocessing of text data
```

```python
import string
```

```python
# normalize the text/ preprocessing
X = [x.lower() for x in X ] # convert to lower case
```

```python
X
```

```python
X=[''.join(c for c in x if c not in string.punctuation) for x in X]# remove punctuation marks
```

```python
X
```

```python
X=[''.join(c for c in x if c not in '0123456789') for x in X]# remove numbers
```

```python
X=[' '.join(x.split()) for x in X]# REMOVE the extra whitespace
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=100,stratify=Y)
```

```python
X_train
```

```python
Y_train
```

```python
import nltk
```

```python
def tokenzier(text):
    words=nltk.word_tokenize(text)
    return words
```

```python
# creation of Bag of Words
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
#countvec=CountVectorizer(stop_words="english")

countvec=TfidfVectorizer(stop_words=None,norm=None)
```

```python
#Reason of not getting good accuracy (for example TfidfVectorizer)
 #stop_words:- Usually it remove the common words but many a times , these stop_words may carry some important information.
 # normalization: The TfidfVectorizer bydefault it use L2 normalization, it sclae down the values between 0&1.
# if we cahnge it Norm=None, there is no such restrictions. Usually it give improvement.
```

```python
new_X_train=countvec.fit(X_train)
```

```python
X_train_vectorized=countvec.transform(X_train)
```

```python
X_train_vectorized
```

```python
X_test_vectorized=countvec.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(X_train_vectorized,Y_train)
```

```python
prediction=mnb.predict(X_test_vectorized)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(Y_test,prediction)
```

```python
confusion_matrix(Y_test,prediction)
```

```python
print(classification_report(Y_test,prediction))
```

```python
# lets write a function to check new sentence -- category
```

```python
def enter_text(message):
    if mnb.predict(countvec.transform([message]))==0:
        print("the message seems to be neutral")
    elif mnb.predict(countvec.transform([message]))==1:
        print("the message seems to be NEGATIVE REMARK")
    else:
        print("No Problem , it is a POSITIVE REMARK")
```

```python
# reading a specific page from the PDF file
page_2=reader.pages[2]
page2_text=page_2.extract_text()
print(page2_text)
```

```python
import nltk
from nltk import word_tokenize
page2_text_tokens=nltk.word_tokenize(page2_text)
print("word tokonized...\n")
print("#####"*3)
print(page2_text_tokens)
```

```python
# apply stop words

from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words("english"))
```

```python
page2_filtered_text=[word for word in page2_text_tokens if not word.lower() in stop_words]
page2_cleantext=[word.lower() for word in page2_filtered_text if word.isalpha()]
```
